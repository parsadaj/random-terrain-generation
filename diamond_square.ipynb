```python
import numpy as np
```

```python
def printArray(array):
    for i in range(array.shape[0]):
        for j in range(array.shape[1]):
            print(array[i,j], end=' ')
        print()
```

```python
def diamond(array, i, j, array_size):
    array_size = array.shape[0]
    n = 0
    left, right, up, bottom = 0, 0, 0, 0
    if i > 0:
        left  = array[i-1, j]
        n += 1
    if i + 1 < array_size:
        right = array[i+1, j]
        n += 1
    if j > 0:
        up = array[i, j-1]
        n += 1
    if j + 1 < array_size:
        bottom = array[i, j+1]
        n += 1
    
    return (left + right + up + bottom) / n
        
    
        
            
```

```python
n_iter = 10
seed_numbers = [10*np.random.randn() for i in range(4)]

array = np.array([[seed_numbers[0], seed_numbers[1]], [seed_numbers[2], seed_numbers[3]]])
array_size = 2
printArray(array)

for i in range(n_iter):
    

    # expand array
    expand_size = array_size - 1
    new_array_size = expand_size + array_size

    new_array = np.zeros((new_array_size, new_array_size))

    for i in range(array_size):
        for j in range(array_size):
            new_array[2*i, 2*j] = array[i, j]
    
    array = new_array
    array_size = new_array_size
    
    # print()
    # print('----------------------')
    # print()
    
    # printArray(array)
    

    # fill square
    for i in range(1, array_size, 2):
        for j in range(1, array_size, 2):
            array[i, j] = (array[i+1, j+1] + array[i-1, j+1] + array[i+1, j-1] + array[i-1, j-1]) / 4 + np.random.randn()
    # print()
    # print('----------------------')
    # print()
    
    # printArray(array)

    # fill diamond
    for i in range(0, array_size, 2):
        for j in range(1, array_size, 2):
            array[i,j] = diamond(array, i, j, array_size) + np.random.randn()

    for j in range(0, array_size, 2):
        for i in range(1, array_size, 2):
            array[i,j] = diamond(array, i, j, array_size) + np.random.randn()

    # print()
    # print('----------------------')
    # print()
    # printArray(array)
```
